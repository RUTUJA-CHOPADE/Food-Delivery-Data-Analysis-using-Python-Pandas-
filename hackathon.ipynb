```python
restaurant_summary = food_df.groupby('restaurant_name_x').agg(
    total_orders=('order_id','count'),
    avg_order_value=('total_amount','mean')
)
```

```python
small_restaurants = restaurant_summary[restaurant_summary['total_orders'] < 20]
```

```python
top_small_restaurant = small_restaurants['avg_order_value'].idxmax()
top_small_restaurant
```
